```python
def analizar_columna(df: pd.DataFrame, columna: str) -> dict:
     # Función para analizar las estadisitcas basicas de cualquier columna

     if columna not in df.columns:
          return "Columna no encontrada"

     resultados = {}

     resultados['media'] = df[columna].mean()
     resultados['mediana'] = df[columna].median()
     resultados['moda'] = df[columna].mode()[0]
     resultados['desviacion'] = df[columna].std()

     # Asimetría y Curtosis
     resultados['asimetria'] = df[columna].skew()  # 0 Asimetría perfecta, <0 Asimetría negativa, >0 Asimetría positiva 
     resultados['curtosis'] = df[columna].kurtosis()  # Probabilidad de outliers, 0 Curtosis normal, <0 Menos outliers, >0 Mas outliers

     sns.histplot(df[columna], kde=True, bins=30)
     plt.axvline(resultados['mediana'], color='red', linestyle='--', label='Mediana')
     plt.legend()
     plt.show()

     # Interpretación
     if -0.5 < resultados['asimetria'] < 0.5:
          print(f"✅ La asimetría está dentro del rango aceptable para normalidad (±0.5) -> {resultados['asimetria']:.2f}")
     else:
          print(f"❌ La asimetría está fuera del rango aceptable. {'Asimetría positiva (cola derecha)' if resultados['asimetria'] > 0 else 'Asimetría negativa (cola izquierda)'} -> {resultados['asimetria']:.2f}")

     if -0.5 < resultados['curtosis'] < 0.5:
          print(f"✅ La curtosis está dentro del rango aceptable para normalidad {resultados['curtosis']:.2f}")
     else:
          print(f"❌ La curtosis está fuera del rango aceptable. {'Leptocúrtica (más outliers)' if resultados['curtosis'] > 0 else 'Platicúrtica (menos outliers)'} -> {resultados['curtosis']:.2f}")

analizar_columna(df=df, columna='Quantity')
```

```python
from scipy import stats

def colum_normal(df : pd.DataFrame, columna: str) -> None:
   #Función para realizar tests de normalidad en una columna específica del DataFrame
   muestra = df[columna].dropna()

   # Para Shapiro-Wilk, que es sensible a muestras muy grandes (>5000), 
   # tomamos una submuestra aleatoria si es necesario. Es una buena práctica.
   muestra_shapiro = muestra.copy()
   if len(muestra_shapiro) > 5000:
      muestra_shapiro = muestra_shapiro.sample(5000, random_state=666)
   
   # PRUEBAS ESTADÍSTICAS

   # --- Shapiro-Wilk Test ---
   # Compara los cuantiles de los datos con los de una distribución normal.
   # Usa la muestra de tamaño controlado.
   shapiro_test = stats.shapiro(muestra_shapiro)
   
   # --- Kolmogorov-Smirnov Test ---
   # Compara la distribución de los datos con una normal ESTÁNDAR (media=0, std=1).
   # Por lo tanto, PRIMERO DEBEMOS ESTANDARIZAR nuestros datos.
   datos_estandarizados = (muestra - muestra.mean()) / muestra.std()
   ks_test = stats.kstest(datos_estandarizados, 'norm')
   
   # --- D'Agostino's K^2 Test ---
   # Evalúa si la asimetría y la curtosis de los datos son similares a las de una dist. normal.
   k2_test = stats.normaltest(muestra)

   # INTERPRETACIÓN DE RESULTADOS
   # Usamos un nivel de significancia (alpha) de 0.05.
   # Hipótesis Nula (H0): Los datos provienen de una distribución normal.
   # Si p-valor < alpha, rechazamos H0.
   print(f"Tests de normalidad para '{columna}':")
   print(f"Shapiro-Wilk test: estadístico={shapiro_test[0]:.4f}, p-value={shapiro_test[1]:.6f}")
   print(f"Kolmogorov-Smirnov test: estadístico={ks_test[0]:.4f}, p-value={ks_test[1]:.6f}")
   print(f"D'Agostino's K^2 test: estadístico={k2_test[0]:.4f}, p-value={k2_test[1]:.6f}")

   alpha = 0.05
   # La lógica aquí es conservadora: solo si TODAS las pruebas fallan en rechazar H0,
   # consideramos que los datos podrían ser normales.
   if shapiro_test[1] > alpha and ks_test[1] > alpha and k2_test[1] > alpha:
      print("\n✅ Todos los tests sugieren que la distribución podría ser normal.")
   else:
      print("\n❌ Al menos un test sugiere que la distribución NO es normal.")

colum_normal(df, "Quantity")
```

📊 Aplicamos tres pruebas estadísticas:

# Shapiro-Wilk (precisa en muestras pequeñas)
# Kolmogorov-Smirnov (útil en muestras grandes)
# D'Agostino's K² (asimetría y curtosis combinadas)
# Esto nos da un panorama más robusto para decidir si la distribución es realmente normal.

```python
def evaluar_normalidad_visual(df, columna):
   """
   Evalúa visualmente si una columna sigue una distribución normal 
   mediante tres gráficos clave.
   """
   plt.figure(figsize=(18, 5))
   
   # --- Gráfico 1: Histograma y Curva de Densidad ---
   plt.subplot(1, 3, 1)
   # QUÉ BUSCAR: Que la forma general se parezca a la clásica "campana de Gauss", simétrica.
   sns.histplot(df[columna], kde=True, color='blue')
   plt.title(f'Histograma de {columna}')
   
   # --- Gráfico 2: Q-Q Plot (Quantile-Quantile) ---
   plt.subplot(1, 3, 2)
   # Esta es la prueba visual más importante. Compara los cuantiles de tus datos
   # con los cuantiles teóricos de una distribución normal.
   # QUÉ BUSCAR: Que los puntos azules caigan perfectamente sobre la línea roja diagonal.
   # Las desviaciones, especialmente en los extremos, indican que no es normal.
   stats.probplot(df[columna].dropna(), plot=plt)
   plt.title(f'Q-Q Plot de {columna}')
   
   # --- Gráfico 3: Boxplot (Diagrama de Caja) ---
   plt.subplot(1, 3, 3)
   # QUÉ BUSCAR:
   # 1. Simetría: La línea de la mediana debe estar en el centro.
   # 2. Outliers: Que no haya (o haya muy pocos) puntos fuera de los "bigotes".
   sns.boxplot(y=df.query(f"{columna} < 500")[columna], color='green')
   plt.title(f'Boxplot de {columna}')
   plt.tight_layout()
   plt.show()
   
# Llamada a la función para ejecutar el análisis visual
evaluar_normalidad_visual(df, "Quantity")
```

```python
# Correlacion entre 2 variables categoricas
# Se usa chi-cuadrado para evaluar la independencia entre dos variables categóricas
from scipy.stats import chi2_contingency

# Aleatorizar y preparar los datos
np.random.seed(42)
observacion = pd.crosstab(df["2urvived"], df["Sex"], margins=True)
observacion
```

```python
# Prueba Chi-cuadrado
test = chi2_contingency(observacion)
chi = test[0]  # Estadístico Chi-cuadrado (h0 son independientes)
p = test[1]   # Valor p (h1 es dependiente)
print(f"Estadístico Chi-cuadrado: {chi:.4f}")
print(f"Valor p: {p:.4f}")
# Interpretación
alpha = 0.05
if p < alpha:
   print("❌ Rechazamos la hipótesis nula: las variables son dependientes.")
   print("Conclusión: Las variables SON dependientes. Hay una relación estadísticamente significativa entre el género y la supervivencia.")
else:
   print("✅ No rechazamos la hipótesis nula: las variables son independientes.")
   print("Conclusión: Las variables SON independientes. No hay evidencia de una relación entre el género y la supervivencia.")
```

```python
# Correlacion entre dos variables numericas

coeficiente, p_valor = stats.pearsonr(df['Age'], df['Fare']) # Tambien se puede usar corr() de pandas
print(f"Coeficiente de correlación de Pearson: {coeficiente:.4f}") # H0 es que no hay correlación
print(f"Valor p: {p_valor:.4f}") # H1 es que hay correlación
if p_valor < 0.05:
   print("\nConclusión: La correlación es estadísticamente significativa.")
   print("Por ende hay relacion entre la edad y la tarifa pagada.")
else:
   print("\nConclusión: La correlación NO es estadísticamente significativa.")
   print("Por ende no hay relacion entre la edad y la tarifa pagada.")
```

```python
# Correlación entre variable numerica y categórica
# Se usa ANOVA (3 o mas grupos) y Test t (2 grupos)

# Separamos las edades en dos grupos según el sexo
edad_hombres = df[df['Sex'] == 0]['Age']
edad_mujeres = df[df['Sex'] == 1]['Age']

# Realizamos el Test t para muestras independientes
t_stat, p_valor = stats.ttest_ind(edad_hombres, edad_mujeres)
print(f"Estadístico t: {t_stat:.2f}") # H0 es que las medias son iguales
print(f"Valor p: {p_valor:.4f}") # H1 es que las medias son diferentes
alpha = 0.05
if p_valor < alpha:
    print("❌ Rechazamos la hipótesis nula: las medias de edad entre hombres y mujeres son diferentes.")
else:
    print("✅ No rechazamos la hipótesis nula: las medias de edad entre hombres y mujeres son iguales.")
```

```python
# ANOVA Queremos saber si la edad promedio de los pasajeros era diferente según su clase social.

#Hipótesis Nula (H₀): La edad media de las 3 clases es igual.
#Hipótesis Alternativa (H₁): Al menos una de las clases tiene una edad media diferente a las demás

edad_clase1 = df[df['Pclass'] == 1]['Age']
edad_clase2 = df[df['Pclass'] == 2]['Age']
edad_clase3 = df[df['Pclass'] == 3]['Age']
# Realizamos el test ANOVA de una vía
f_stat, p_valor = stats.f_oneway(edad_clase1, edad_clase2, edad_clase3)

print(f"Estadístico F: {f_stat:.2f}")
print(f"P-valor: {p_valor}")

# Interpretamos el resultado
if p_valor < 0.05:
    print("\nConclusión: El p-valor es menor que 0.05. Rechazamos la hipótesis nula.")
    print("Existe una diferencia estadísticamente significativa en la edad promedio entre al menos dos de las clases.")
else:
    print("\nConclusión: No hay evidencia para rechazar la hipótesis nula.")
```

```python
import statsmodels.stats.multicomp as mc
posthoc_result = mc.pairwise_tukeyhsd(endog=df['Age'], 
                                      groups=df['Pclass'], 
                                      alpha=0.05)

# 2. Mostramos los resultados en una tabla clara
print("\n--- Prueba Post-Hoc de Tukey HSD ---")
print(posthoc_result) # True -> La diferencia de edad promedio es significativa.
```
