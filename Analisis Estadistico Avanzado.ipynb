```python
def analizar_columna(df: pd.DataFrame, columna: str) -> dict:
     # FunciÃ³n para analizar las estadisitcas basicas de cualquier columna

     if columna not in df.columns:
          return "Columna no encontrada"

     resultados = {}

     resultados['media'] = df[columna].mean()
     resultados['mediana'] = df[columna].median()
     resultados['moda'] = df[columna].mode()[0]
     resultados['desviacion'] = df[columna].std()

     # AsimetrÃ­a y Curtosis
     resultados['asimetria'] = df[columna].skew()  # 0 AsimetrÃ­a perfecta, <0 AsimetrÃ­a negativa, >0 AsimetrÃ­a positiva 
     resultados['curtosis'] = df[columna].kurtosis()  # Probabilidad de outliers, 0 Curtosis normal, <0 Menos outliers, >0 Mas outliers

     sns.histplot(df[columna], kde=True, bins=30)
     plt.axvline(resultados['mediana'], color='red', linestyle='--', label='Mediana')
     plt.legend()
     plt.show()

     # InterpretaciÃ³n
     if -0.5 < resultados['asimetria'] < 0.5:
          print(f"âœ… La asimetrÃ­a estÃ¡ dentro del rango aceptable para normalidad (Â±0.5) -> {resultados['asimetria']:.2f}")
     else:
          print(f"âŒ La asimetrÃ­a estÃ¡ fuera del rango aceptable. {'AsimetrÃ­a positiva (cola derecha)' if resultados['asimetria'] > 0 else 'AsimetrÃ­a negativa (cola izquierda)'} -> {resultados['asimetria']:.2f}")

     if -0.5 < resultados['curtosis'] < 0.5:
          print(f"âœ… La curtosis estÃ¡ dentro del rango aceptable para normalidad {resultados['curtosis']:.2f}")
     else:
          print(f"âŒ La curtosis estÃ¡ fuera del rango aceptable. {'LeptocÃºrtica (mÃ¡s outliers)' if resultados['curtosis'] > 0 else 'PlaticÃºrtica (menos outliers)'} -> {resultados['curtosis']:.2f}")

analizar_columna(df=df, columna='Quantity')
```

```python
from scipy import stats

def colum_normal(df : pd.DataFrame, columna: str) -> None:
   #FunciÃ³n para realizar tests de normalidad en una columna especÃ­fica del DataFrame
   muestra = df[columna].dropna()

   # Para Shapiro-Wilk, que es sensible a muestras muy grandes (>5000), 
   # tomamos una submuestra aleatoria si es necesario. Es una buena prÃ¡ctica.
   muestra_shapiro = muestra.copy()
   if len(muestra_shapiro) > 5000:
      muestra_shapiro = muestra_shapiro.sample(5000, random_state=666)
   
   # PRUEBAS ESTADÃSTICAS

   # --- Shapiro-Wilk Test ---
   # Compara los cuantiles de los datos con los de una distribuciÃ³n normal.
   # Usa la muestra de tamaÃ±o controlado.
   shapiro_test = stats.shapiro(muestra_shapiro)
   
   # --- Kolmogorov-Smirnov Test ---
   # Compara la distribuciÃ³n de los datos con una normal ESTÃNDAR (media=0, std=1).
   # Por lo tanto, PRIMERO DEBEMOS ESTANDARIZAR nuestros datos.
   datos_estandarizados = (muestra - muestra.mean()) / muestra.std()
   ks_test = stats.kstest(datos_estandarizados, 'norm')
   
   # --- D'Agostino's K^2 Test ---
   # EvalÃºa si la asimetrÃ­a y la curtosis de los datos son similares a las de una dist. normal.
   k2_test = stats.normaltest(muestra)

   # INTERPRETACIÃ“N DE RESULTADOS
   # Usamos un nivel de significancia (alpha) de 0.05.
   # HipÃ³tesis Nula (H0): Los datos provienen de una distribuciÃ³n normal.
   # Si p-valor < alpha, rechazamos H0.
   print(f"Tests de normalidad para '{columna}':")
   print(f"Shapiro-Wilk test: estadÃ­stico={shapiro_test[0]:.4f}, p-value={shapiro_test[1]:.6f}")
   print(f"Kolmogorov-Smirnov test: estadÃ­stico={ks_test[0]:.4f}, p-value={ks_test[1]:.6f}")
   print(f"D'Agostino's K^2 test: estadÃ­stico={k2_test[0]:.4f}, p-value={k2_test[1]:.6f}")

   alpha = 0.05
   # La lÃ³gica aquÃ­ es conservadora: solo si TODAS las pruebas fallan en rechazar H0,
   # consideramos que los datos podrÃ­an ser normales.
   if shapiro_test[1] > alpha and ks_test[1] > alpha and k2_test[1] > alpha:
      print("\nâœ… Todos los tests sugieren que la distribuciÃ³n podrÃ­a ser normal.")
   else:
      print("\nâŒ Al menos un test sugiere que la distribuciÃ³n NO es normal.")

colum_normal(df, "Quantity")
```

ðŸ“Š Aplicamos tres pruebas estadÃ­sticas:

# Shapiro-Wilk (precisa en muestras pequeÃ±as)
# Kolmogorov-Smirnov (Ãºtil en muestras grandes)
# D'Agostino's KÂ² (asimetrÃ­a y curtosis combinadas)
# Esto nos da un panorama mÃ¡s robusto para decidir si la distribuciÃ³n es realmente normal.

```python
def evaluar_normalidad_visual(df, columna):
   """
   EvalÃºa visualmente si una columna sigue una distribuciÃ³n normal 
   mediante tres grÃ¡ficos clave.
   """
   plt.figure(figsize=(18, 5))
   
   # --- GrÃ¡fico 1: Histograma y Curva de Densidad ---
   plt.subplot(1, 3, 1)
   # QUÃ‰ BUSCAR: Que la forma general se parezca a la clÃ¡sica "campana de Gauss", simÃ©trica.
   sns.histplot(df[columna], kde=True, color='blue')
   plt.title(f'Histograma de {columna}')
   
   # --- GrÃ¡fico 2: Q-Q Plot (Quantile-Quantile) ---
   plt.subplot(1, 3, 2)
   # Esta es la prueba visual mÃ¡s importante. Compara los cuantiles de tus datos
   # con los cuantiles teÃ³ricos de una distribuciÃ³n normal.
   # QUÃ‰ BUSCAR: Que los puntos azules caigan perfectamente sobre la lÃ­nea roja diagonal.
   # Las desviaciones, especialmente en los extremos, indican que no es normal.
   stats.probplot(df[columna].dropna(), plot=plt)
   plt.title(f'Q-Q Plot de {columna}')
   
   # --- GrÃ¡fico 3: Boxplot (Diagrama de Caja) ---
   plt.subplot(1, 3, 3)
   # QUÃ‰ BUSCAR:
   # 1. SimetrÃ­a: La lÃ­nea de la mediana debe estar en el centro.
   # 2. Outliers: Que no haya (o haya muy pocos) puntos fuera de los "bigotes".
   sns.boxplot(y=df.query(f"{columna} < 500")[columna], color='green')
   plt.title(f'Boxplot de {columna}')
   plt.tight_layout()
   plt.show()
   
# Llamada a la funciÃ³n para ejecutar el anÃ¡lisis visual
evaluar_normalidad_visual(df, "Quantity")
```

```python
# Correlacion entre 2 variables categoricas
# Se usa chi-cuadrado para evaluar la independencia entre dos variables categÃ³ricas
from scipy.stats import chi2_contingency

# Aleatorizar y preparar los datos
np.random.seed(42)
observacion = pd.crosstab(df["2urvived"], df["Sex"], margins=True)
observacion
```

```python
# Prueba Chi-cuadrado
test = chi2_contingency(observacion)
chi = test[0]  # EstadÃ­stico Chi-cuadrado (h0 son independientes)
p = test[1]   # Valor p (h1 es dependiente)
print(f"EstadÃ­stico Chi-cuadrado: {chi:.4f}")
print(f"Valor p: {p:.4f}")
# InterpretaciÃ³n
alpha = 0.05
if p < alpha:
   print("âŒ Rechazamos la hipÃ³tesis nula: las variables son dependientes.")
   print("ConclusiÃ³n: Las variables SON dependientes. Hay una relaciÃ³n estadÃ­sticamente significativa entre el gÃ©nero y la supervivencia.")
else:
   print("âœ… No rechazamos la hipÃ³tesis nula: las variables son independientes.")
   print("ConclusiÃ³n: Las variables SON independientes. No hay evidencia de una relaciÃ³n entre el gÃ©nero y la supervivencia.")
```

```python
# Correlacion entre dos variables numericas

coeficiente, p_valor = stats.pearsonr(df['Age'], df['Fare']) # Tambien se puede usar corr() de pandas
print(f"Coeficiente de correlaciÃ³n de Pearson: {coeficiente:.4f}") # H0 es que no hay correlaciÃ³n
print(f"Valor p: {p_valor:.4f}") # H1 es que hay correlaciÃ³n
if p_valor < 0.05:
   print("\nConclusiÃ³n: La correlaciÃ³n es estadÃ­sticamente significativa.")
   print("Por ende hay relacion entre la edad y la tarifa pagada.")
else:
   print("\nConclusiÃ³n: La correlaciÃ³n NO es estadÃ­sticamente significativa.")
   print("Por ende no hay relacion entre la edad y la tarifa pagada.")
```

```python
# CorrelaciÃ³n entre variable numerica y categÃ³rica
# Se usa ANOVA (3 o mas grupos) y Test t (2 grupos)

# Separamos las edades en dos grupos segÃºn el sexo
edad_hombres = df[df['Sex'] == 0]['Age']
edad_mujeres = df[df['Sex'] == 1]['Age']

# Realizamos el Test t para muestras independientes
t_stat, p_valor = stats.ttest_ind(edad_hombres, edad_mujeres)
print(f"EstadÃ­stico t: {t_stat:.2f}") # H0 es que las medias son iguales
print(f"Valor p: {p_valor:.4f}") # H1 es que las medias son diferentes
alpha = 0.05
if p_valor < alpha:
    print("âŒ Rechazamos la hipÃ³tesis nula: las medias de edad entre hombres y mujeres son diferentes.")
else:
    print("âœ… No rechazamos la hipÃ³tesis nula: las medias de edad entre hombres y mujeres son iguales.")
```

```python
# ANOVA Queremos saber si la edad promedio de los pasajeros era diferente segÃºn su clase social.

#HipÃ³tesis Nula (Hâ‚€): La edad media de las 3 clases es igual.
#HipÃ³tesis Alternativa (Hâ‚): Al menos una de las clases tiene una edad media diferente a las demÃ¡s

edad_clase1 = df[df['Pclass'] == 1]['Age']
edad_clase2 = df[df['Pclass'] == 2]['Age']
edad_clase3 = df[df['Pclass'] == 3]['Age']
# Realizamos el test ANOVA de una vÃ­a
f_stat, p_valor = stats.f_oneway(edad_clase1, edad_clase2, edad_clase3)

print(f"EstadÃ­stico F: {f_stat:.2f}")
print(f"P-valor: {p_valor}")

# Interpretamos el resultado
if p_valor < 0.05:
    print("\nConclusiÃ³n: El p-valor es menor que 0.05. Rechazamos la hipÃ³tesis nula.")
    print("Existe una diferencia estadÃ­sticamente significativa en la edad promedio entre al menos dos de las clases.")
else:
    print("\nConclusiÃ³n: No hay evidencia para rechazar la hipÃ³tesis nula.")
```

```python
import statsmodels.stats.multicomp as mc
posthoc_result = mc.pairwise_tukeyhsd(endog=df['Age'], 
                                      groups=df['Pclass'], 
                                      alpha=0.05)

# 2. Mostramos los resultados en una tabla clara
print("\n--- Prueba Post-Hoc de Tukey HSD ---")
print(posthoc_result) # True -> La diferencia de edad promedio es significativa.
```
